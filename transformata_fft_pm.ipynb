Wygeneruj próbkę sygnału cyfrowego y=sin( 200 *pi*t)+cos( 300 *pi*t) i oblicz dla niego transformatę Fouriera (może być szybka transformata Fouriera). Narysuj wykres modułu i argumentu transformaty. Co udało się zauważyć? Dodaj teraz do próbkowanych wartości sygnału liczby losowe z przedziału [-1,1] i jeszcze raz narysuj wykres sygnału, policz transformatę i narysuj wykresy jej modułu i argumentu. Co się zmieniło?

```python
import numpy as np
from scipy.fft import fft,fftfreq #części fftpack i signal z biblioteki SciPy
import matplotlib.pyplot as plt
#dyskretną transformatę Fouriera sekwencji próbek obliczamy za pomocą algorytmu szybkiej transformaty Fouriera (FFT)

def FFTclear(t,y,N,Tp): #FFT = Fastest Fourier Transform
    yf=fft(y)
    xf=fftfreq(N,Tp)[:N//2] #ffftfreq to f-cja obliczająca częstotliwości odpowiadające poszczególnym wart. z wyniku otrzymanego z wywołania fft
    plt.plot(xf, 2.0/N * np.abs(yf[0:N//2])) #wykres modułu i argumentu transformaty
    #[:N//2] bo interesuje nas tylko częstotliwość dodatnia, ujemna będzie jej odbiciem lustrzanym 
def FFTrand(t,y,N,Tp): 
    yr=rand(y)
    yf=fft(yr)
    xf=fftfreq(N,Tp)[:N//2] 
    plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))   
def rand(X):
    return X+np.random.uniform(low=-1,high=1) #dodajemy wartości z [-1,1]
rand=np.vectorize(rand)


fp=2000 #częstotliwość próbkowania
Tp=1/fp #okres próbkowania

N=300 #ilość próbek

t=np.linspace(0,N*Tp,N,endpoint=False)
y=np.sin(200*2*np.pi*t)+np.cos(300*2*np.pi*t)

#wykresy
plt.subplot(2,1,1)
FFTclear(t,y,N,Tp)
plt.subplot(2,1,2)
FFTrand(t,y,N,Tp)
plt.show()
#wykres 1: widzimy dwa widma sygnału
#wykres 2: widzimy, że dodane wartości [-1,1] z rand powodują zakłócenia i widać bardzo małe widma częstotliwościowe, które się nakładają na siebie
```
