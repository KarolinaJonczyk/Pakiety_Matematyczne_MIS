Zadanie $1$

Narysuj wykresy funkcji $y = sin(x), y = cos(x), y = tg(x), y = ctg(x)
$w przedziale $[−π, π]$ tak, aby nadawaay się do kart wzorów.
Nie zapomnij o porządnym dobraniu zmienności $x$ oraz opisaniu wykresów 
(tytuł, osie). Wyeksportuj je do plików grafcznych i wykorzystaj w dokumencie $LaTeX$.

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi,500)
plt.plot(x, np.sin(x))


plt.title(r'$\sin(x)$') #tytuł wykresu
ax = plt.gca()
ax.set_aspect(2.0)
#siatka
plt.grid(color='grey', linestyle='-', linewidth=0.5) 

#osie OX i OY
ax.axhline(0, color='black',linewidth=1)
ax.axvline(0, color='black', linewidth=1)


#opis OX
plt.xticks([-math.pi, (-1/2)*math.pi, (-1/3)*math.pi, (-1/6)*math.pi, 0, (1/6)*math.pi,
            (1/4)*math.pi, (1/3)*math.pi, (1/2)*math.pi, math.pi], ['$- \\pi$','$- \\frac{\\pi}{2}$',
            '$- \\frac{\\pi}{3}$', '$- \\frac{\\pi}{6}$', '$0$', '$\\frac{\\pi}{6}$', '$\\frac{\\pi}{4}$',
            '$\\frac{\\pi}{3}$', '$\\frac{\\pi}{2}$', '$\\pi$'])


#strzałki dla osi
ax.plot((1), (0), ls="", marker=">", ms=5, color="k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot((0), (1), ls="", marker="^", ms=5, color="k", transform=ax.get_xaxis_transform(), clip_on=False)



#ukrywanie ramki
for line in ["left", "right", "bottom", "top"]:
    ax.spines[line].set_visible(False)

#miejsce wartości na osi
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')


#opis osi na strzałkach
plt.xlabel("x", loc="right")
plt.ylabel("y", loc="top")


print("Wykres funkcji sin(x) ma postać: ")
plt.show()
```

```python
x = np.linspace(-1 * math.pi, math.pi, 1000)
y = np.tan(x)


# NaN - "Not a Number"
y[:-1][np.diff(y) < 0] = np.nan
# show grid
ax = plt.gca()
plt.grid(color='grey', linestyle='-', linewidth=0.5) 

#osie OX i OY
ax.axhline(0, color='black',linewidth=1)
ax.axvline(0, color='black', linewidth=1)

#strzałki dla osi
ax.plot((1), (0), ls="", marker=">", ms=5, color="k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot((0), (1), ls="", marker="^", ms=5, color="k", transform=ax.get_xaxis_transform(), clip_on=False)


#ukrywanie ramki
for line in ["left", "right", "bottom", "top"]:
    ax.spines[line].set_visible(False)

#miejsce wartości na osi
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')

#opis osi na strzałkach 
plt.xlabel("x", loc="right")
plt.ylabel("y", loc="top")

#usuwanie na osi
plt.ylim(-10,10)
plt.xlim(-1 * math.pi, math.pi)

#przeliczanie na radiany
radian_multiples = [-1, -3/4, -1/2, -1/4, 0, 1/4, 1/2, 3/4, 1]
radians = [n * math.pi for n in radian_multiples]
radian_labels = [ '$-\pi$','$-\\frac{3\\pi}{4}$', '$-\\frac{\\pi}{2}$', '$-\\frac{\\pi}{4}$', '0', '$\\frac{\\pi}{4}$', '$\\frac{\\pi}{2}$','$\\frac{3\\pi}{4}$', '$\pi$']

#asymptoty
ax.axvline((-1/2)*math.pi, color='red', linestyle='dashed')
ax.axvline((1/2)*math.pi, color='red', linestyle='dashed')


plt.xticks(radians, radian_labels)


plt.title("$tan(x)$", fontsize=14)
plt.plot(x, y)
print("Wykres funkcji tan(x) ma postać: ")
plt.show()
```

Zadanie $3$

Narysuj ilustrację do obliczenia pola figury ograniczonej funkcją $y = x^2 - 4x + 1$ oraz prostymi $y=3x -5$ i $y = -2$.

```python
#krzywe
x = np.linspace(0, 10,1000)
y1 = x**2 -4*x + 1
y2 = 3*x - 5
y3 = -2
y1plot = plt.plot(x, x**2 -4*x + 1, 'royalblue', label='$x^2 -4x +1$')
y2plot = plt.plot(x, 3*x - 5, 'hotpink', label='$3x-5$')
y3plot = plt.axhline(y = -2 , color= 'seagreen', label = '$y=-2$')

#wypełnienie powierzchni
plt.fill_between(x, y1, 3*x-5, where=y1<3*x-5, color='peru', alpha = 1)
plt.fill_between(x, y3, x**2 - 4*x + 1, where=y3>x**2 - 4*x + 1, color='white', alpha = 1)


#legenda
plt.legend(loc='lower right')

#siatka
plt.grid(color='grey', linestyle='-', linewidth=0.5) 

print("Oblicz pole zamalowanego obszaru pokazanego na poniższym wykresie: ")
plt.show()
```

Zadanie $4$

Narysuj wykres funkcji dwóch zmiennych $f(x,y) = sin \left(3cos \left(x^2 + y^2 \right) \right)$ na obszarze $\left[ -2, 2 \right] \times \left[-1, 1 \right]$.

```python
x = np.linspace(-2, 2, 20)
y = np.linspace(-1, 1, 20)

X, Y = np.meshgrid(x, y)
Z = np.sin(3 * np.cos ( X**2 + Y**2))

plt.contourf(X, Y, Z, levels=20, cmap='Greens')
plt.colorbar()
plt.show()
```

Zadanie $5$

Narysuj wykres $3D$, na którym będzie paraboloida obrotowa przecięta płaszczyzną.

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.pyplot import cm

x = np.arange(-8, 8,0.1)
y = np.arange(-8, 8,0.2)
X, Y = np.meshgrid(x, y)
Z= X**2 + Y**2
#Z[Z<X+Y]=-8
Z2= 50 + X*0
#Z2[Z>X+Y]=-8

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.view_init(0, 45)
ax.plot_surface(X,Y,Z, cmap= cm.coolwarm, linewidth=0, antialiased=False)
ax.plot_wireframe(X, Y, Z2, cmap='Pastel1', cstride=5, rstride=5) 

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('$x^2+ y^2$')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# .linspace arguments are (start, end, number_of_steps)
x = np.linspace(-2 * math.pi, 2 * math.pi, 1000)
y = np.tan(x)


# This operation inserts a NaN where the difference between successive points is negative
# NaN means "Not a Number" and NaNs are not plotted or connected
# I found this by doing a search for "How to plot tan(x) in matplotlib without the connecting lines between asymtotes"
y[:-1][np.diff(y) < 0] = np.nan
# show grid
plt.grid()
tol = 10
y[y > tol] = np.nan
y[y < -tol] = np.nan


plt.xlabel("x")
plt.ylabel("$tan(x)$")

# Set the x and y axis cutoffs
plt.ylim(-10,10)
plt.xlim(-2 * math.pi, 2 * math.pi)

# x_labels in radians
# For a more programmatic approach to radians, see https://matplotlib.org/3.1.1/gallery/units/radian_demo.html
radian_multiples = [-1, -5/6, -3/4, -7/12, -1/2, -1/3, -1/4, -1/6, 0, 1/6, 1/4, 1/3, 1/2, 7/12, 3/4, 5/6, 1]
radians = [n * math.pi for n in radian_multiples]
radian_labels = ['$-\\pi$', '$-\\frac{5\\pi}{6}$','$-\\frac{3\\pi}{4}$', '$-\\frac{7\\pi}{12}$', '$-\\frac{\\pi}{2}$','$-\\frac{\\pi}{3}$','$-\\frac{\\pi}{4}$','$-\\frac{\\pi}{6}$', '0','$\\frac{\\pi}{6}$', '$\\frac{\\pi}{4}$', '$\\frac{\\pi}{3}$',  '$\\frac{\\pi}{2}$', '$\\frac{7\\pi}{12}$', '$\\frac{3\\pi}{4}$', '$\\frac{5\\pi}{6}$','$\\pi$']

plt.xticks(radians, radian_labels)


plt.title("$y = tan(x)$", fontsize=14)
plt.plot(x, y)
plt.show()
```
