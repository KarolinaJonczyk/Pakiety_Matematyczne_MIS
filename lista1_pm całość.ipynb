Zadanie $3$

Dla $x \in \left< -100, 100 \right>$, oblicz wartość przybliżeń <br>
$ e_N (x) =  \sum\limits_{k=0}^{N} \frac{x^k}{k!}$ <br> 
dla $N = 100$ oraz $N = 1000$ i porównaj z wartościmi funkcji $h(x) = e^x$.

```python
def suma(N):
    s = 0
    while s <= N:
        s += 1
    print(s) 
```

```python
def silnia(k):
    if k==0:
        return 1
    else:
        licz = k*silnia(k-1)
        return licz
```

```python
print(silnia(5)) #sprawdzenie czy działa silnia
```

```python
#dwie pętle, najpierw dla N = 100
```

```python
X = []

def podszereg(n):
    for x in range (-100, 101):
        k = 0
        for k in range (0,n+1):
            suma = (x**k)/silnia(k)
            X.append(suma)
            k += 1
    return X
    
```

```python
podszereg(100)
```

```python
sigma1 = sum(podszereg(100))

print(sigma1)
```

```python
podszereg(1000)
```

```python
sigma2= sum(podszereg(1000))

print(sigma2)
```

```python
H = [] 
for x in range (-100, 101):
    h= math.exp(x)
    H.append(h)
```

```python
print(H)
```

```python
exp = sum(H)

print(exp)
```

```python
wynik1 = abs(exp - sigma1)

print("Różnica między h(x)=e^x , a szeregiem z zadania dla N=100, to: ", wynik1)

wynik2 = abs(exp-sigma2)

print("\n", "Różnica między h(x)=e^x , a szeregiem z zadania dla N=1000, to: ", wynik2)
```

Zadanie $5$

Rozwiąż układu równań liniowych:

$$\left\{\begin{array}{rcl}
2x + y + z&=&1\\
x-y-2z&=&3\\
x + y + 3z&=& 10,
\end{array} \right.$$

$$\left\{\begin{array}{rcl}
x + y + z&=&0\\
x + 3y + 2z&=&1\\
2x + 5y + 6z&=& 3 \\
4x + 2y + z &=&4 \\
x + 6y - 5z &=& 1.
\end{array} \right.$$

Rozwiązanie:

Zamieniam układ równań na macierz dopełnień

```python
def cramer(a, b):
    if np.linalg.det(a) == 0:
        return "Podany układ równań nie ma rozwiązania!"
    else:
        return np.linalg.solve(a, b)
```

```python
up = np.array([[2, 1, 1], [1, -1, -2], [1, 1, 3]])
upans = np.array([1,3,10])
print(up, "\n" ,upans)
```

```python
upsolutions = cramer(up, upans)
```

```python
print(upsolutions)
```
