```python
g1(x)=(x-1)^4
```

```python
g2(x)=x^4-4*x^3+6*x^2-4*x+1
```

```python
g1(1-1/1000)
```

```python
g2(1-1/1000)
```

```python
g1(1-1/1000)-g2(1-1/1000)
```

```python
import numpy as np
x=np.linspace(1-1/1000,1+1/1000,100)
```

```python
[g1(a) for a in x]==[g2(a) for a in x]
```

```python
a=1-1/1000
b=1+1/1000
h=(b-a)/100
x1=[a+i*h for i in range(1,101)]
[g1(a) for a in x1]==[g2(a) for a in x1]
```

### Wykresy 3D

```python
x, y = var('x y')
W = plot3d(x^2+y^2,(x,-2,2),(y,-2,2), color='purple', opacity=0.8)
S = plot3d(x-y+2,(x,-2,2),(y,-2,2), color='green',opacity=0.8)
show(W)
show(W + S,viewer='tachyon')
show(W + S,viewer='threejs')
```
