```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
x = np.linspace(1/(2*math.e), math.e**2+1)
y = np.log(x)

plt.plot(x, y)
plt.show()
```

```python
y2 = np.log(1/x)

#plt.plot(x,y,x,y2,label=['1','2'])  # legenda nie dzia≈Ça
plt.plot(x, y, label='ln(x)')
plt.plot(x, y2, label='$\\ln(\\frac{1}{x})$')
plt.legend(loc='right')
plt.show()
```

```python
plt.plot(x, y, 'r>', linestyle='--', label='ln(x)')
plt.plot(x, y2, 'g3',label='ln(1/x)')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
plt.plot(x, y, label='ln(x)')
plt.plot(x, y2,label='$\\ln(\\frac{1}{x})$')
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Przyk≈Çad')
plt.grid()
plt.xticks([0, 1/math.e,1, math.e,2*math.e,math.e**2],\
          ['$0$','$\\frac{1}{e}$','$1$', '$e$','$2e$','$e^2$'])
plt.yticks([-2,-1,0,1,2])
plt.show()
```

```python
x = np.linspace(1/(2*math.e), math.e**2+1)
y = np.log(x)
y2 = np.log(2*x)
y3 = np.log(1/x)

plt.subplot(2, 3, 1) # upper left
plt.plot(x, y, label='ln(x)')
plt.plot(x, y2, label='ln(2x)')
plt.legend(loc='best')

plt.subplot(2, 3, 2) # upper right
plt.plot(x, y, label='ln(x)')
plt.plot(x, y3,label='ln(1/x)')
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Przyk≈Çad')
plt.grid(True)
plt.xticks([1/math.e,1,math.e,2*math.e,math.e**2],\
          ['$\\frac{1}{e}$','$1$', '$e$','$2e$','$e^2$'])
plt.yticks([-2,-1,0,1,2])

x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)


plt.subplot(2, 3, 3) # bottom left
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2 


plt.contourf(X, Y, Z, levels=20, cmap='Reds')
plt.colorbar()

plt.subplot(2, 3, 4)
Z2 = X**2 - y**2

plt.contourf(X, Y, Z, levels=20)
plt.colorbar()

plt.savefig('matplotlib_example.jpg')

plt.show()
```

### Histogram

```python
h = np.random.normal(1, 2, 1000)

plt.subplot(1,2,1)
n, bins, patches = plt.hist(h, bins=30)

plt.subplot(1, 2, 2)
plt.hist(h, bins=20, histtype='bar', edgecolor='k', density=True)
plt.suptitle('Histogram')
plt.show()
```

```python
x = np.random.normal(0,2,100)
y = np.random.normal(0,2,100)

plt.figure()
plt.hist2d(x, y, bins=(20, 30), cmap='BuGn')
plt.colorbar()
plt.title('hist2d plot')

plt.figure()
plt.hexbin(x, y,bins=20, gridsize=20, extent=(-5, 5, -5, 5), cmap='Reds')
plt.colorbar()
plt.title('hexbin plot')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(-4, 4,0.1)
y = np.arange(-4, 4,0.2)
X, Y = np.meshgrid(x, y)
Z=np.sin(X)+np.cos(Y)
#Z[Z<X+Y]=-8
Z2=X+Y
#Z2[Z>X+Y]=-8

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.view_init(45, -48)
ax.plot_wireframe(X,Y,Z2)
#ax.plot_surface(X, Y, Z, cmap='plasma')
#ax.plot_surface(X,Y,Z2)
ax.plot_wireframe(X, Y, Z, cmap='plasma')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('$\sin(x)+\cos(y)$')
plt.show()
```
